```python
%matplotlib inline

from azureml import Workspace
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ws = Workspace()
ds = ws.datasets['bcpdata.csv']
bc_data = ds.to_dataframe()

print(bc_data)
```

```python
#dropping unwanted columns

bc_data = bc_data.drop(bc_data.columns[[0]], axis = 1)

print(bc_data) 
```

```python


#removing rows with missing values

bc_data.dropna(axis = 0, inplace =True)



```

```python
#pair-wise scatter plot,The red dots correspond to malignant diagnosis and blue to benign
import seaborn as sns

features_mean= list(bc_data.columns[0:10])

color_dic = {'M':'red', 'B':'blue'}
colors = bc_data['diagnosis'].map(lambda x: color_dic.get(x))

sm = pd.scatter_matrix(bc_data[features_mean], c=colors, alpha=0.4, figsize=((15,15)));





```

```python
#box plot

sns.boxplot(x='diagnosis', y='area_mean', data=bc_data)
```
